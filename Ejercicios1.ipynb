```python
import ROOT
f = ROOT.TFile.Open("Tracks_Clusters.root")
canvas = ROOT.TCanvas("Canvas","a first way to plot a variable",800,600)
tree = f.Get("JetRecoTree")
tree.Print()
tree.GetEntries()
```

# Ejercicio 1

```python
hist_1 = ROOT.TH1F("mu_average","average interactions per bunch-crossing",80,0,80)
```

```python
for event in tree:
    hist_1.Fill(tree.mu_average)
print("Done!")
```

```python
hist_1.SetFillColor(2)
hist_1.Draw()
canvas.Draw()
```

# Ejercicio 5

## Tracks:

### pT : Momentum trasverso asociado a cada track.
### eta : Variable asociada al angulo polar theta del sistema de referencia del detector y sus valores dependen de la posicion y configuración del tracker.
### phi : Angulo polar que junto a eta permite determinar la linea de trayectoria.
### m : Masa registrada, en este caso no tiene significado fisico. 
### vtx : numero de tracks por vertice común.


## Clusters:

### pT : Momentum trasverso asociado a cada cluster.
### eta : Variable asociada al angulo polar theta del sistema de referencia del detector y sus valores dependen de la posicion y configuración de los diferentes calorimetros.
### phi : Angulo polar que junto a eta permite determinar la trayectoria del cluster.
### m : La "masa" registrada del cluster. Da cero debido a que el cluster es medido en el calorimetro y en este el cluster es absorbido totalmente como energía entonces se le asocia masa 0.


## Particles:

### pT : El momentum trasverso de la particula, esta es la componente del momentum en la dirección de haz. 
### eta : Variable asociada al angulo polar theta del sistema de referencia del detector, en este caso es el valor simulado por monte carlo entonces presenta una distribución más uniforme que la de los demás.
### phi : Angulo polar que junto a eta permite determinar la trayectoria de la partícula.
### m : La masa de cada particula registrada.

### PDG ID: El Particle Data Group PDG asigna un ID o numero caracteristico a cada particula conocida e hipotetica y a este se le llama PDG ID, esto significa que un histograma de las PDG ID de un experimento muestra el numero de cada una de las particulas conocidas e hipoteticas que se registró. Los valores negativos corresponden a la antiparticula de cada partícula a la cual se le asocia un PDG ID positivo.
